```python
import img2pdf
import os
from PIL import Image
```

```python
img_files = os.listdir("data/questions_image")
for file in img_files:
    file_name = file.split(".")[0]
    with open(f"data/questions_pdf/{file_name}.pdf", "wb") as f:
        f.write(img2pdf.convert([f"data/questions_image/{file}"]))
```
