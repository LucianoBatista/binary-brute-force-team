```python
import img2pdf
import os
from PIL import Image

from tqdm import tqdm
```

```python
img_files = os.listdir("data/questions_image_with_description")
for file in tqdm(img_files):
    file_name = file.split(".")[0]
    with open(f"data/questions_pdf_with_description/{file_name}.pdf", "wb") as f:
        f.write(img2pdf.convert([f"data/questions_image_with_description/{file}"]))
```
